```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
df = pd.read_csv('processed_energy_data.csv')
```

```python
df.head()
```

```python
def get_real_time_tariff(hour):
    if 0 <= hour < 6:
        return 1.0  # Off-peak rate
    elif 6 <= hour < 12:
        return 2.5  # Mid-rate
    elif 12 <= hour < 18:
        return 3.5  # Peak rate
    else:
        return 4.0  # High-demand rate
```

```python
# Apply the real-time tariff function
df['tariff_rate'] = df['hour'].apply(get_real_time_tariff)
```

```python
# Function to simulate real-time monitoring of data
def simulate_real_time_data(df, start_index=0, interval=60):
    current_time_index = start_index
    while current_time_index < len(df):
        # Select the row corresponding to the current index
        current_row = df.iloc[current_time_index]

        # Display the real-time tariff rate
        current_time = current_row['time']
        current_usage = current_row['use [kW]']
        current_generation = current_row['gen [kW]']
        current_tariff_rate = current_row['tariff_rate']

        print(f"Time: {current_time}, Usage: {current_usage} kW, Generation: {current_generation} kW, Tariff Rate: Rs {current_tariff_rate} / kWh")

        # Sleep to simulate real-time monitoring (adjust the interval for faster or slower simulation)
        time.sleep(1)

        current_time_index += interval  # Update the index by 'interval' rows for each iteration
```

```python
# Step 5: Plot a visualization of tariff rates over time
def plot_tariff_rates(df):
    plt.figure(figsize=(12, 6))
    plt.plot(df['time'], df['tariff_rate'], label='Tariff Rate (Rs/kWh)', color='purple')
    plt.title('Real-Time Tariff Monitoring')
    plt.xlabel('Time')
    plt.ylabel('Tariff Rate (Rs/kWh)')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# Simulate real-time tariff monitoring
print("Starting Real-Time Tariff Simulation...")
simulate_real_time_data(df.head(1000), start_index=0, interval=60)
```
